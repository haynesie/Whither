```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
data = pd.read_csv('allconsolelogins.csv')
```

```python
hourseries = data['hour']
```

```python
xlabels = ['9:00am', '10:00am', '11:00am', '12:00pm', '1:00pm', '2:00pm', '3:00pm', '4:00pm', '5:00pm', '6:00pm', '7:00pm', '8:00pm']
```

```python
hourcounts = hourseries.value_counts(sort=False)
```

```python
plot1 = hourcounts.plot.bar(color='m' ,title="Console Logins by Hour", rot=0, figsize=(11,6))
plot1.set(xlabel='Hour', ylabel='Logins', xticklabels=xlabels)
```

```python
dateseries = data['date']
```

```python
datecounts = dateseries.value_counts(sort=False)
```

```python
ax= datecounts.plot.bar(color='c', title='Console logins by Date in February', rot=0, figsize=(10,6))
ax.set(xlabel='date', ylabel='number of logins')
```
