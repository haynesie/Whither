```python
import itertools
```

```python
columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
rows = [ 1, 2, 3, 4]
```

```python
def plotMatrix(matrix, col_labels, row_labels, title='Matrix Heatmap', cmap=plt.cm.Greys):
    threshold = matrix.max()/2
    plt.imshow(matrix, cmap)
    plt.title(title)
    plt.colorbar()
    plt.yticks(np.arange(len(row_labels)), row_labels)
    plt.xticks(np.arange(len(col_labels)), col_labels)
    fmt = '.2f'
    for i, j in itertools.product(range(matrix.shape[0]), range(matrix.shape[1])):
        plt.text(j, i, format(matrix[i, j], fmt), horizontalalignment="center", color = "black" \
        if matrix[i,j] < threshold else "white")
    plt.show()
```

```python
SysLogins = pd.read_csv('Data/LoginsByMachine.csv')
```

```python
SysLogins.describe()
```

```python
logins = SysLogins['logins'].get_values()
```

```python
loginmatrix = np.matrix(logins.reshape(4,8))
```

```python
plotMatrix(loginmatrix, columns, rows, title='Workstation Logins Heatmap', cmap=plt.cm.Blues)
```

```python
Crashes = pd.read_csv('Data/CrashesByMachine.csv')
```

```python
Crashes.describe()
```

```python
crashmatrix = np.matrix(Crashes['crashes'].get_values().reshape(4,8))
```

```python
plotMatrix(crashmatrix, columns, rows, title='System Crashes by Machine', cmap=plt.cm.Reds)
```
